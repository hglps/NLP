```python
import spacy
import nltk
import os
nltk.download()

import nlpnet
workpath = os.path.dirname(os.path.abspath('spacytests.ipynb'))
c = os.path.join(workpath, 'pos-pt')
```

```python
tagger = nlpnet.POSTagger(c, language='pt')

print(tagger.tag('olá meu nome é'))

nlp = spacy.load("pt_core_news_sm")
doc = nlp("Esta é uma frase.")
print([(w.text, w.pos_) for w in doc])
```
